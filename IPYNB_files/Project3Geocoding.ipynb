```python
file_in = '../clean_data/clean_home_no_dupes.csv'
```

```python
# Dependencies and Setup
from ast import Param
import pandas as pd
import requests

dataframe = pd.read_csv(file_in)
dataframe.head()
```

```python
index = dataframe.index.to_list()
results_list = []
API_KEY = "get_your_own"


lat_list = []
lng_list = []
for i in index:
    # row = row of index value, i
    row = dataframe.iloc[[i]]
    row_dict = row.to_dict()
    
    
    address_entry = {}
    
    address = row_dict['full_address'][i]

    params = {
        'key': API_KEY,
        'address' : address
    }

    base_url = "https://maps.googleapis.com/maps/api/geocode/json?"

    response = requests.get(base_url, params=params).json()
    response.keys()

    if response['status'] == "OK":
        geometry = response['results'][0]["geometry"]
        lat = geometry['location']['lat']
        lon = geometry['location']['lng']


        address_entry['lat'] = lat
        address_entry['lng'] = lon
    
        lat_list.append(lat)
        lng_list.append(lon)
    results_list.append(address_entry)
    print('-----------------------')
    print(f'processing row {i}')
    print('-----------------------')
```

```python
dataframe['lat'] = lat_list
dataframe['lng'] = lng_list
x = dataframe['lat'].nunique()
x
```

```python
dropped = dataframe.dropna()
dropped
```
